```python
file_paths=[f"wasserstein/results_ot_{i}.csv" for i in range(10)]

dfs = []

for file_path in file_paths:
    df = pd.read_csv(file_path)

    # Convert string representations of lists to actual lists
    df["BASETEXTO"] = df["BASETEXTO"].apply(ast.literal_eval)
    df["RANDOMPATCH"] = df["RANDOMPATCH"].apply(ast.literal_eval)
    df["NNPROJ"] = df["NNPROJ"].apply(ast.literal_eval)
    df["AFFINE"] = df["AFFINE"].apply(ast.literal_eval)

    dfs.append(df)

# Define a custom function to calculate the mean and std for lists
def mean_list(series):mean_std_ot
    return list(np.mean(series.tolist(), axis=0))

def std_list(series):
    return list(np.std(series.tolist(), axis=0))

result_df = pd.concat(dfs).groupby(level=0).agg({
    'BASETEXTO': [mean_list, std_list],
    'RANDOMPATCH': [mean_list, std_list],
    'NNPROJ': [mean_list, std_list],
    'AFFINE': [mean_list, std_list],
})

result_df.columns = ['_'.join(col).strip() for col in result_df.columns.values]
```

```python
result_df.to_csv("mean_std_ot.csv", index=False)
```

```python
def calculate_mean(lst):
    return mean(lst)

# Apply the function to each column with "_list" in its name
for col in result_df.columns:
    if '_list' in col:
        result_df[col] = result_df[col].apply(calculate_mean)
```
